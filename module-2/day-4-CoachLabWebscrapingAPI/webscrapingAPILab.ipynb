```python
all_quotes = []
for x in range(1, 11):
    new_url = f"{url}/page/{x}"
#     print('NEW URL HERE', new_url)
    dummy_url = requests.get(new_url)
#     print('DUMMY URL',dummy_url)
    dummy_soup = BeautifulSoup(dummy_url.text)
    quotes = dummy_soup.findAll('div', {'class': 'quote'})
    print('QUOTES',quotes)
    print('LENGTH OF QUOTES',len(quotes))
    all_quotes.append(quotes)
```

```python
len(all_quotes)
```

```python
len(all_quotes[0])
```

```python
flat_list = []
for sublist in all_quotes:
    print('this is the sublist',sublist)
    for item in sublist:
        flat_list.append(item)
print('length of flat list',len(flat_list))
```

```python
flat_list[0]
```

```python
flat_list[0].find('small', {'class': 'author'}).text
```

```python
quotes_dictionary = {}
for x in flat_list:
    quote = x.find('span', {'class': 'text'}).text
    author = x.find('small', {'class' : 'author'}).text
    quotes_dictionary[len(quotes_dictionary)] = [quote,author]
```

_id, name, species, wikilink

```python
elephant_dictionary = {}

for x in elephant_json:  
    elephant_id = x['_id']    
    
    try:
        name = x['name']
        species = x['species']
        wikilink = x['wikilink']
    except:
        print('Missing Value')
    elephant_dictionary[len(elephant_dictionary)] = [elephant_id,name,species,wikilink]
```

```python
df = pd.DataFrame(elephant_dictionary).T
df.columns = ['_id', 'name', 'species', 'wikilink']
df.head()
```
